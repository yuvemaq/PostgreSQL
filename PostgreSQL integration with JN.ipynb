## Intermediate

3. We are launching a platinum service for the loyalest of the loyal customers. We will assign platinum status to customers that have had 40 or more transaction payments. What customer_ids are eligible for platinum status?

```python
%%sql
SELECT customer_id, COUNT(amount) FROM payment
GROUP BY customer_id
HAVING COUNT(amount) >=40;
```

4. What are the customer ids of customers who have spent more than $100$ in payment transactions with our staff_id member 2?

```python
%%sql
SELECT customer_id, SUM(amount) FROM payment
WHERE staff_id = 2
GROUP BY customer_id
HAVING SUM(amount) > 100;
```

5. Inner join customer and payment tables

```python
%%sql
SELECT DISTINCT(customer.customer_id), first_name, last_name, email, amount, payment_date
FROM customer
INNER JOIN payment
ON payment.customer_id = customer.customer_id
LIMIT 20;
```

6. Outer join customer and payment tables

```python
%%sql
SELECT * FROM customer
FULL OUTER JOIN payment
ON customer.customer_id = payment.customer_id
LIMIT 20;
```

```python
%%sql
SELECT * FROM customer
FULL OUTER JOIN payment
ON customer.customer_id = payment.customer_id
WHERE customer.customer_id IS null OR payment.payment_id IS null;
```

Check work

```python
%%sql
SELECT COUNT(DISTINCT customer_id) FROM customer
```

7. Left outer join payment and customer

```python
%%sql
SELECT * FROM payment
LEFT OUTER JOIN customer
ON payment.customer_id = customer.customer_id
LIMIT 20;
```

8. Grab entries unique to payment

```python
%%sql
SELECT * FROM payment
LEFT OUTER JOIN customer
ON payment.customer_id = customer.customer_id
WHERE customer.activebool IS null
LIMIT 20;
```

9. Find quarter of payment.

```python
%%sql
SELECT EXTRACT(QUARTER FROM payment_date) AS month
FROM payment
LIMIT 20;
```

10. Find how old date stamp is from our current date.

```python
%%sql
SELECT AGE(payment_date)
FROM payment
LIMIT 20;
```

11. In the form of text, find the day, month, date, and year of each payment.

```python
%%sql
SELECT TO_CHAR(payment_date, 'day, mon/d/yyyy')
FROM payment
LIMIT 20;
```

12.-13. During which months did payments occur? How many payments occurred on a Monday? 

```python
%%sql
SELECT DISTINCT(TO_CHAR(payment_date, 'MON'))
FROM payment
LIMIT 20;
```

```python
%%sql
SELECT COUNT(*)
FROM payment
WHERE EXTRACT(dow FROM payment_date) = 1;
```

Here, we won't see any customers that did not make any payments, since we filtered it by joining with payment table.

# Save results as Dataframe

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyplot
%matplotlib inline
```

```python
# Open engine in context manager
# Perform query and save results as df
with engine.connect() as con:
    rs = con.execute('SELECT customer_id, SUM(amount) FROM payment WHERE staff_id = 2 GROUP BY customer_id HAVING SUM(amount) > 100;')
    df = pd.DataFrame(rs.fetchmany(size=5))

# Print length of dataframe
print(len(df))

print(df.head())
```

```python
df.columns = rs.keys()
```
