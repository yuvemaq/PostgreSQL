## Intermediate

3. We are launching a platinum service for the loyalest of the loyal customers. We will assign platinum status to customers that have had 40 or more transaction payments. What customer_ids are eligible for platinum status?

```python
%%sql
SELECT customer_id, COUNT(amount) FROM payment
GROUP BY customer_id
HAVING COUNT(amount) >=40;
```

4. What are the customer ids of customers who have spent more than $100$ in payment transactions with our staff_id member 2?

```python
%%sql
SELECT customer_id, SUM(amount) FROM payment
WHERE staff_id = 2
GROUP BY customer_id
HAVING SUM(amount) > 100;
```

5. Inner join customer and payment tables (limit to 20).

```python
%%sql
SELECT DISTINCT(customer.customer_id), first_name, last_name, email, amount, payment_date
FROM customer
INNER JOIN payment
ON payment.customer_id = customer.customer_id
LIMIT 20;
```

17. Create a custom email using intial of first name, full last name, and domain name gmail.com from customer table (limit to 20).

```python
%%sql
SELECT LOWER(LEFT(first_name, 1)) || LOWER(last_name) || '@gmail.com'
AS custom_email
FROM customer
LIMIT 20;
```

18. Return a list of all the films that were returned between 5/27/2005 to 5/28/2005. 

```python
%%sql
SELECT film_id, title
FROM film
WHERE film_id IN
(SELECT inventory.film_id
FROM rental
INNER JOIN inventory ON inventory.inventory_id = rental.inventory_id
WHERE rental.return_date BETWEEN '2005-05-27' AND '2005-05-28')
ORDER BY title
```

For our subquery, we grab film_id from inventory as it is not available in the rental table, perform an inner join using inventory_id, then filter based on return_date. Then, we can grab the title of the film by using title from film table and use the IN function to include the subquery. To organize this better, outside the subquery ORDER BY title. 

19. Find out the names of the customers who made a payment of greater than 10 dollars 
(limit to 20 names).

```python
%%sql
SELECT first_name, last_name
FROM customer
WHERE EXISTS
(SELECT * FROM payment
WHERE payment.customer_id = customer.customer_id
AND amount > 10)
LIMIT 20;
```

20. Find all pairs of films with same length (limit to 20).

```python
%%sql
SELECT title, length FROM film
WHERE length = 50
```

We want 'Adaptation Holes' to match with 'Blues Instinct' and with the rest of film titles that all have the same length. Ideally, we want three columns in our final table - film1, film2, and length. Again, limit to 20.

```python
%%sql
SELECT filmA.title AS film1, filmB.title AS film2, filmA.length AS length
FROM film AS filmA
INNER JOIN film AS filmB ON
filmA.film_id != filmB.film_id
AND filmA.length = filmB.length
LIMIT 20;
```

# Save results as Dataframe

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyplot
%matplotlib inline
```

```python
# Open engine in context manager
# Perform query and save results as df
with engine.connect() as con:
    rs = con.execute('SELECT filmA.title AS film1, filmB.title AS film2, filmA.length AS length FROM film AS filmA INNER JOIN film AS filmB ON filmA.film_id != filmB.film_id AND filmA.length = filmB.length')
    df = pd.DataFrame(rs.fetchmany(size=20))

# Print length of dataframe
print(len(df))

print(df.head())
```
